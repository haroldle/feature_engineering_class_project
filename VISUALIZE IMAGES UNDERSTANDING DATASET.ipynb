```python
import cv2
from imutils import paths
import numpy as np
import matplotlib.pyplot as plt
from LBP_extractor import extract
from skimage.feature import local_binary_pattern
```

# VISUALIZING THE IMAGES

### GRABBING AN INSTANCE OF EACH CATEGORY

```python
x = paths.list_images('data')
images = {}
data = []
for i in x:
    if i.split('/')[1] not in images.keys():
        images[i.split('/')[1]] = True
        data.append(i)
```

### VISUALIZING THE EACH INSTANCE

```python
w = 15
h = 15
fig = plt.figure(figsize=(w, h))

columns = 8
rows = 6
for i in range(0,  columns*rows - 2):
    img = cv2.imread(data[i])
    fig.add_subplot(rows, columns, i + 1)
    plt.tick_params(left = False, 
                    right = False , 
                    labelleft = False ,
                    labelbottom = False, 
                    bottom = False)
    plt.imshow(img)
plt.show()
```

# VISUALIZING EACH INSTANCE OF EACH CATEGORY IN GRAY SCALE

```python
w = 15
h = 15
fig = plt.figure(figsize=(w, h))

columns = 8
rows = 6
for i in range(0,  columns*rows - 2):
    img = cv2.imread(data[i], 0)
    fig.add_subplot(rows, columns, i + 1)
    plt.tick_params(left = False, 
                    right = False , 
                    labelleft = False ,
                    labelbottom = False, 
                    bottom = False)
    plt.imshow(img, cmap="gray")
plt.show()
```

# VISUALIZING A LBP KERNEL (PREPROCESSING IMAGES)

### LBP UNIFORM AND ROTATION INVARIANT

```python
an_image = cv2.imread(data[0], 0)
lbp_image = local_binary_pattern(an_image, 8, 1.0, 'uniform')
lbp_image_2 = local_binary_pattern(an_image, 16, 2.0, 'uniform')
lbp_image_3 = local_binary_pattern(an_image, 24, 3.0, 'uniform')

columns = 4
rows = 1
fig = plt.figure(figsize=(w, h))

print("FROM LEFT TO RIGHT")
print("ORIGINAL IMAGE => LBP 8 POINTS WITH RADIUS OF 1 => LBP 16 POINTS WITH RADIUS OF 2 => LBP 24 POINTS WITH RADIUS OF 3")

fig.add_subplot(rows, columns, 1)
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)
plt.imshow(an_image, cmap='gray')

fig.add_subplot(rows, columns, 2)
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)
plt.imshow(lbp_image, cmap='gray')

fig.add_subplot(rows, columns, 3)
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)
plt.imshow(lbp_image_2, cmap='gray')

fig.add_subplot(rows, columns, 4)
plt.tick_params(left = False, right = False , labelleft = False ,labelbottom = False, bottom = False)
plt.imshow(lbp_image_3, cmap='gray')
```

### VISUALIZING THE HISTOGRAM OF EACH LBP EXTRACTOR

```python
print("LBP 8 POINTS: ", extract(an_image, 8, 1.0, 'uniform'))
print("LBP 16 POINTS: ", extract(an_image, 16, 2.0, 'uniform'))
print("LBP 24 POINTS: ", extract(an_image, 24, 3.0, 'uniform'))
```

```python
plt.hist(extract(an_image, 8, 1.0, 'uniform'))
plt.show()
plt.hist(extract(an_image, 16, 2.0, 'uniform'))
plt.show()
plt.hist(extract(an_image, 24, 3.0, 'uniform'))
plt.show()
```
